# Creating a Bipartite graph 

```python
import networkx as nx
import matplotlib.pyplot as plt
import math
%matplotlib inline
import pandas as pd
from networkx.algorithms import bipartite
import numpy as np
```

```python
#Creating a Data Frame
df = pd.read_excel(r'D:\Notes\VI SEM\SNA\Source Data.xlsx')
df2= pd.DataFrame(df,columns=["Name","How much were you able to concentrate on your daily activities after workout?"])
df.head(10)
#print(df2)
```

```python
G = nx.Graph()
for i in range(len(df2)) :
     G.add_node(df2.loc[i, "Name"])  

for i in range(len(df2)) :
    G.add_node(df2.loc[i, "How much were you able to concentrate on your daily activities after workout?"])
    
for i in range(len(df2)):
    G.add_edge(df2.loc[i, "Name"],df2.loc[i, "How much were you able to concentrate on your daily activities after workout?"])
```

```python
Age_Concen = nx.from_pandas_edgelist(df,source="Name",target="How much were you able to concentrate on your daily activities after workout?")
```

```python
pos = {}
Names ={}
Concen_level = {}
for x in df["Name"]:
  if x in Names:
    Names[x]+=1
  else:
    Names[x]=1
    pos[x]=(15,(len(Names)+1)*10)
for x in df["How much were you able to concentrate on your daily activities after workout?"]:
  if x in Concen_level:
    Concen_level[x]+=1
  else:
    Concen_level[x]=1
    pos[x]=(30,(len(Concen_level)+1)*60+60)
print(pos)
```

```python
plt.figure(figsize=(20,30))
nx.draw(Age_Concen,with_labels=True,node_color='green',pos=pos)
plt.show()
```

```python
print(nx.info(G))
print(nx.is_directed(G))
```

```python
print('Degree Centrality:',nx.degree_centrality(G))
diction1 = nx.degree_centrality(G)
keymax = max(diction1, key=diction1.get)
print ('Max Degree Centrality Value:',keymax)
```

```python
print('Density of the graph is:',nx.density(G))
```

```python
print("Clustering:", nx.clustering(G).items())
print('Average Clustering is:',nx.average_clustering(G))
```

```python
print('Diameter of the network:',nx.degree(G))
diction2 = dict(nx.degree(G))
keymax2 = max(diction2, key=diction2.get)
print('Maximum Diameter:',keymax2)
```
